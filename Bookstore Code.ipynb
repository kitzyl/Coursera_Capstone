# Part I: Data from Foursquare

We implement our Foursquare API to retrieve a list of bookstores or bookshops within a 7.5km radius of a given suburb and state. We scrape the results for the suburb, postcode, latitude, longitude, and name of each bookshop.

```python
!pip install bs4
#!pip install requests

from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
LIMIT = 50 # limit of number of venues returned by Foursquare API
CLIENT_ID = 'NPSZSY5GS5WFYYY4YGPOEBYF0MJYRXL0N0XIWTTYATQIIF0Y'
CLIENT_SECRET = 'OZLAA5VXSDBM0KOLVRLAAZEZC5NQUS4AZG2ZVJ4K3YVF5I2U'
VERSION = '20180605'
CATEGORY = 'bookshop%2C%20bookstore'
RADIUS = 7500

def getBookstores(suburb,state):
  
    venues_list=[]
            
        # create the API request URL
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&near={},{}&query={}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            suburb, 
            state, 
            CATEGORY,
            RADIUS,
            LIMIT)
            
        # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        
    for v in results:
        #known_suburb=float('NaN')
        #known_postcode=float('NaN')
        if 'city' in v['venue']['location'] and 'postalCode' in v['venue']['location']:
       #     known_suburb = v['venue']['location']['city']
        #if 'postalCode' in v['venue']['location']:
        #    known_postcode = v['venue']['location']['postalCode']
        
            venues_list.append([(
                v['venue']['location']['city'], 
                int(v['venue']['location']['postalCode']),
                v['venue']['location']['lat'],
                v['venue']['location']['lng'],
                v['venue']['name'])])

    bookstores = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    bookstores.columns = ['Suburb', 
                  'Postcode',
                  'Latitude',
                  'Longitude',
                  'Venue']
    
    return(bookstores)
```

We obtain a list of bookstore within 7.5km of Melbourne, Victoria and Sydney, NSW. We print out the Melbourne list as an example.

```python
melb_bookstores = getBookstores("Melbourne","VIC")
sydney_bookstores = getBookstores("Sydney","NSW")

melb_bookstores
#sydney_bookstores
```

# Part III: Combining Foursquare and ATO data

We merge the two dataframes from the above two Parts by their postcodes.

```python
melb_bookstores_bysuburb_income = pd.merge(melb_bookstores_bysuburb,avg_income).reset_index(drop=True)
sydney_bookstores_bysuburb_income = pd.merge(sydney_bookstores_bysuburb,avg_income).reset_index(drop=True)

#sydney_bookstores_bysuburb_income
melb_bookstores_bysuburb_income.head()
```

We then calculate the available income PER bookstore in each suburb.

```python
melb_bookstores_bysuburb_income.loc[:,'Average Income for each Bookstore'] = (melb_bookstores_bysuburb_income['Total Suburb Income']/melb_bookstores_bysuburb_income['# bookstores'])
melb_bookstores_bysuburb_income.head()

sydney_bookstores_bysuburb_income.loc[:,'Average Income for each Bookstore'] = (sydney_bookstores_bysuburb_income['Total Suburb Income']/sydney_bookstores_bysuburb_income['# bookstores'])
#sydney_bookstores_bysuburb_income.head()
```

Then we un-condense the above (tallied) dataframe into the dataframe with individually listed bookstores, while maintaining the income availablt to each bookstore. We do this for both the Melbourne and Sydney bookstores.

```python
available_income =[]
l = melb_bookstores_bysuburb_income.shape[0]
for postcode in melb_bookstores['Postcode']:
    k=0
    for i in range(0,l):
        if melb_bookstores_bysuburb_income['Postcode'][i] == postcode:
            available_income.append(melb_bookstores_bysuburb_income['Average Income for each Bookstore'][i])
            k=1
    if k==0:
        available_income.append(0)

melb_bookstores['Available Income'] = available_income
melb_bookstores.head()
```

```python
available_income_syd =[]
m = sydney_bookstores_bysuburb_income.shape[0]
for postcode in sydney_bookstores['Postcode']:
    k=0
    for i in range(0,m):
        if sydney_bookstores_bysuburb_income['Postcode'][i] == postcode:
            available_income_syd.append(sydney_bookstores_bysuburb_income['Average Income for each Bookstore'][i])
            k=1
    if k==0:
        available_income_syd.append(0)

sydney_bookstores['Available Income'] = available_income_syd
```

# Part IV: Maps
    
We use the above data to map where existing bookstores are in Melbourne, as well as Sydney. We group the bookstores by similar available income (i.e. potential customers) as represented by similar colours.

```python
#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
!pip install folium
import folium # map rendering library
```

```python
#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import numpy as np
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

We include a column called pct_rank for our dataframes, which standardises the Available Income column to a number between 0 and 1. This is to make it easier to colour-code the points on the map as the rgb2hex only takes a value between 0 and 1.

```python
melb_bookstores['pct_rank'] = melb_bookstores['Available Income'].rank(pct=True) #(method = 'max')
melb_bookstores.head()

sydney_bookstores['pct_rank'] = sydney_bookstores['Available Income'].rank(pct=True) #(method = 'max')
#sydney_bookstores.head()
```

The following code generates a map of the Melbourne bookstores.

```python
address = 'Melbourne, Victoria'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude

# create Melbourne map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)


# add markers to the map
markers_colors = []
for lat, lon, label, col in zip(melb_bookstores['Latitude'], melb_bookstores['Longitude'], melb_bookstores['Venue'], melb_bookstores['pct_rank']):
  #  label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=7,
        popup=label,
        #color=rainbow[int(col)-1],
        #fill=True,
        #fill_color=rainbow[int(col)-1],
        color='white',#colors.rgb2hex((col,0.5,0.5)),
        fill=True,
        fill_color=colors.rgb2hex((col,0.75,0.7)),
        fill_opacity=1).add_to(map_clusters)
       
map_clusters
```

```python
print('There are ', sum(melb_bookstores_bysuburb_income['# bookstores']), 'bookstores altogether within 7.5km of ', address,'. The suburbs that the bookstores are in have a combined income of', '${:,.2f}'.format(sum(melb_bookstores_bysuburb_income['Total Suburb Income'])),'from ','{:,}'.format(sum(melb_bookstores_bysuburb_income['Number of individuals lodging an income tax return'])),'individuals.')
```
